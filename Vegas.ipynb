```scala
import java.io._

//plot.json

def renderHTML = {
  val file = new File("Plot.html")
  val bw = new BufferedWriter(new FileWriter(file))
  bw.write(plot.html.frameHTML("foo"))
  bw.close()  
  //plot.html.pageHTML("foo") // a complete HTML page containing the plot
  //plot.html.frameHTML("foo") // an iframe containing the plot
}

def renderJson = {
  val file = new File("Plot.json")
  val bw = new BufferedWriter(new FileWriter(file))
  bw.write(plot.asCirceJson.toString)
  bw.close()  
  //plot.html.pageHTML("foo") // a complete HTML page containing the plot
  //plot.html.frameHTML("foo") // an iframe containing the plot
}
```
